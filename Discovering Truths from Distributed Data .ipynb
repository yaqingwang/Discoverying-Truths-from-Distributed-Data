```python
def get_obj(object_source_claim, obj_list, mean, percent, source_weight, pool, voting):
    #source_weight = {}
    var_dict = {}
    answer_matrix = {}
    
    for obj in object_source_claim:
        n = 0
        var = 0
        right = 0
        num = 0
        cat = True
        truth_weight = {}
        answer_matrix[obj] = [0] * 10 #create answer matrix
        
        for u in object_source_claim[obj]:
            if True:
                claim = object_source_claim[obj][u]
                if u not in source_weight:
                    source_weight[u] = 1
                
                var_list = []
                if not voting:
                    cat = False
                    var += math.pow(claim - mean[obj],2) * 1
                    n += 1
                else:
                        
                    answer_matrix[obj][int(claim-1)] += source_weight[u]

                    n += source_weight[u]
                    if claim not in truth_weight:
                        truth_weight[claim] = 1
                    else:
                        truth_weight[claim] += 1
        if voting:          
            answer_matrix[obj] = [a/(n ) for a in answer_matrix[obj]]
        if cat:
            answer = [0] * 10
            answer[int(mean[obj] - 1)] = 1
            
            var_dict[obj] = distance(answer, answer_matrix[obj])
            if n == 1:
                var_dict[obj] = 1
      
        else:
      
            var_dict[obj] = round(var / chi2.ppf(0.025, n),10)
           
            if n == 1:
                var_dict[obj] = 10
    
    sorted_x = sorted(var_dict.items(), key=operator.itemgetter(1))
    new_obj = []
    size = int(len(sorted_x))
    
    for l in sorted_x[:size]:
        if l[1] >= 0:
            new_obj.append(l[0])
   
    #print var_dict
    return var_dict, answer_matrix

```

```python
def distance(vector_1, vector_2):
    d = 0
    for i in range(len(vector_1)):
        d += math.pow(vector_1[i] - vector_2[i], 2)
    return d
    
```

# source to truth

```python
def source_to_truth(source_weight, object_source_claim, voting):
      #print source_weight
    mean = {}
    store_dict = {}
    #for obj in object_source_claim:
    for obj in object_source_claim:
        for u in object_source_claim[obj]:
            claim = object_source_claim[obj][u]
            #if claim > 6:
                #print claim
            if u not in source_weight:
                source_weight[u] = 0.0001
            if obj in store_dict and claim in store_dict[obj]:
                store_dict[obj][claim] = store_dict[obj][claim] + source_weight[u]
            else:
                if obj not in store_dict:
                    store_dict[obj] = {}
                    store_dict[obj][claim] =source_weight[u]
                elif claim not in store_dict[obj]:
                    store_dict[obj][claim] = source_weight[u]
    #print store_dict
    # categorical:
    if voting:
        print("voting")
        #print store_dict
        #for obj in object_source_claim:
        for obj in object_source_claim:
            mean[obj] = max(store_dict[obj].iteritems(), key=operator.itemgetter(1))[0]
            
    else:
        for obj in object_source_claim:
            m = 0.0
            sum_weight = 0.0001
            if min(list(store_dict[obj].values())) < 0:
                mean[obj] = max(store_dict[obj].iteritems(), key=operator.itemgetter(1))[0]
                continue
            for source in object_source_claim[obj]:
                if source not in source_weight:
                    continue
                #print math.exp(math.pow(object_source_claim[obj][source] - mean[obj], 2))
                w = source_weight[source]
                sum_weight += w
                m += object_source_claim[obj][source] * w
                
            my_data = list(store_dict[obj].keys())
            my_weights = list(store_dict[obj].values())
            mean[obj] = m / sum_weight
            #mean[obj] = ws.weighted_median(my_data, weights=my_weights)
    #print source_weight
    for obj in object_source_claim:
        v = 0.0
        sum_weight = 0.1
        for source in object_source_claim[obj]:
            if source not in source_weight:
                continue
            w = source_weight[source]
            sum_weight += w
            v += math.pow((object_source_claim[obj][source] - mean[obj]), 2) * w
        var[obj]  = v / sum_weight
    return mean, var
```

# Update Weight

```python
def update_weight(source_object_distribution, estimated_truth, obj_var, log_l, object_list):
    source_data = {} # 
    whole = 0
    for s in source_object_distribution:
        if s not in source_data:
            source_data[s] = 1
        #for l in source_object_distribution[s]:
        for l in source_object_distribution[s]:
            #if l in object_list:
            #print estimated_truth[l] - source_object_distribution[s][l][0]
            if not log_l:
                if l not in estimated_truth:
                    print estimated_truth
                    print l
                    break
                source_data[s] += source_object_distribution[s][l][1] + \
                math.pow(source_object_distribution[s][l][0] - estimated_truth[l], 2)
            else:                 
                source_data[s] += source_object_distribution[s][l][1] + \
                math.pow(source_object_distribution[s][l][0] - estimated_truth[l], 2) / (2 * obj_var[l] + delta)
                source_data[s] += math.log(math.sqrt(obj_var[l]) + delta)
                source_data[s] += math.log(2 * math.pi)
        whole += source_data[s]
    for s in source_data:
        source_data[s] = -math.log(source_data[s] / whole)
    return source_data
```

```python
def model(maxIter, new_source_object_claim, new_object_source_claim, obj_list, mean, percent, \
                                        source_weight, object_pool, voting, i):
    source_weight = {}
    percent = 0.8
    for it in range(maxIter):
        source_weight, score = estimate_source(new_source_object_claim, new_object_source_claim, obj_list, mean, percent, \
                                        source_weight, object_pool, voting)
        new_mean, var = source_to_truth(source_weight, new_object_source_claim, voting)
        mean = new_mean
    return new_mean, var, score
```

```python
def computation(source_object_distribution, object_source_distribution, maxIter, delta, loge_like, voting):
    
    source_weight = {}
    for l in source_object_distribution:
        var = 0.0
        n_o = 0.0
        for o in source_object_distribution[l]:
            n_o += 1
            var += source_object_distribution[l][o][1]
        var = var / n_o
        source_weight[l] = 1
    estimated_truth = {}
    iter = 0
    object_list = {}
    while iter < maxIter*10:
        size = 10000
       
        last_source_weight = copy.deepcopy(source_weight)
        
        estimated_truth = {}
        for l in object_source_distribution:
            truth_weight = {}
            if l not in estimated_truth:
                estimated_truth[l] = 0
            for u in object_source_distribution[l]:
                mean = object_source_distribution[l][u][0]
                #print "mean is " + str(mean)  
                if mean not in truth_weight:
                    truth_weight[mean] = 0
                truth_weight[mean] +=  source_weight[u]
           
            if voting:
                estimated_truth[l] = int(max(truth_weight.iteritems(), key=operator.itemgetter(1))[0])
                
            else:
                weight_sum = 0.0
                truth_sum = 0.0
                for truth in truth_weight:
                    truth_sum += truth * truth_weight[truth]
                    weight_sum += truth_weight[truth]
                estimated_truth[l] = truth_sum / weight_sum
                my_data = list(truth_weight.keys())
                my_weights = list(truth_weight.values())

        if log_like:
            obj_var = get_object_var(source_object_distribution, object_source_distribution, source_weight, estimated_truth)
            source_weight = update_weight(source_object_distribution, estimated_truth, obj_var, log_like, object_list)
            
        else:
            source_weight = update_weight(source_object_distribution, estimated_truth, None, log_like, object_list)
       
        if differenceOfWeights(source_weight, last_source_weight) < 0.1 * delta:
            break
        iter += 1
    return estimated_truth
```
